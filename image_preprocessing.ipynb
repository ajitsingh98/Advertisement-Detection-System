```python
import os
import glob
import shutil
import tkinter as tk
from PIL import ImageTk, Image
```

```python
#
image_dir = "classification_data/stirring/"

# Create a directory to move the images
move_dir = "classification_data/nothing_1/"
os.makedirs(move_dir, exist_ok=True)

# Get a list of image file paths in the directory
image_files = glob.glob(os.path.join(image_dir, "*.jpg"))  # Update the file extension as per your image format
```

```python
image_files
```

```python
def open_image():
    global current_index, image_files

    if current_index < len(image_files):
        image_path = image_files[current_index]

        # Open the image using PIL
        image = Image.open(image_path)
        image = image.resize((300, 300))  # Resize the image if needed
        photo = ImageTk.PhotoImage(image)

        # Update the image in the GUI
        image_label.config(image=photo)
        image_label.image = photo
```

```python
def delete_image():
    global current_index, image_files

    if current_index < len(image_files):
        image_path = image_files[current_index]

        # Delete the image file
        os.remove(image_path)
        print("Image deleted.")

        # Remove the image path from the list
        image_files.pop(current_index)

        # Open the next image
        open_image()

def move_image():
    global current_index, image_files

    if current_index < len(image_files):
        image_path = image_files[current_index]

        # Move the image file to the designated folder
        shutil.move(image_path, move_dir)
        print("Image moved.")

        # Remove the image path from the list
        image_files.pop(current_index)

        # Open the next image
        open_image()
```

```python
def next_image():
    
    global current_index, image_files

    current_index += 1

    if current_index < len(image_files):
        open_image()
    else:
        image_label.config(image="")
        image_label.image = None
        
# Create the GUI window
window = tk.Tk()
window.title("Image Inspector")

# Create the image label
image_label = tk.Label(window)
image_label.pack()

# Create the buttons
delete_button = tk.Button(window, text="Delete", command=delete_image)
delete_button.pack(side=tk.LEFT, padx=10)

move_button = tk.Button(window, text="Move", command=move_image)
move_button.pack(side=tk.LEFT, padx=10)

next_button = tk.Button(window, text="Next", command=next_image)
next_button.pack(side=tk.LEFT, padx=10)

# Initialize the current image index
current_index = 0

# Open the first image
open_image()

# Start the GUI event loop
window.mainloop()
```

```python
def merge_files(source_folder, destination_folder):
    
    for filename in os.listdir(source_folder):
        source_file = os.path.join(source_folder, filename)
        destination_file = os.path.join(destination_folder, filename)

        # Check if the destination file already exists
        if os.path.exists(destination_file):
            # Generate a new unique filename by appending a number
            base_name, extension = os.path.splitext(filename)
            counter = 1
            while os.path.exists(destination_file):
                new_filename = f"{base_name}_{counter}{extension}"
                destination_file = os.path.join(destination_folder, new_filename)
                counter += 1

        # Copy the file to the destination folder
        shutil.copy2(source_file, destination_file)

```

```python
merge_files("classification_data/nothing_1/", "classification_data/nothing/")
```
